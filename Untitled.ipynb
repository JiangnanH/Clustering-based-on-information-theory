```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fileSP_si = open('./dataset/SP500/sp500_SI.d')

dataSP_si= []


for line in fileSP_si:
    line = line.strip().split(' ')
    dataSP_si.append(list(float(line[i]) for i in range(len(line))))

dataSP_si = np.array(dataSP_si)
```

```python
T_list = [5,10,15,20,25,30,35,40]
f,ax = plt.subplots(1,2,figsize=(12,4))
plt.suptitle('1/T = [5,10,15,20,25,30,35,40]')

ax[0].plot(dataSP_si[0][8:16],dataSP_si[0][0:8],marker='o',linestyle='--',label='Nc = 5',alpha=0.8)
ax[0].plot(dataSP_si[1][8:16],dataSP_si[1][0:8],marker='o',linestyle='--',label='Nc = 10',alpha=0.8)
ax[0].plot(dataSP_si[2][8:16],dataSP_si[2][0:8],marker='o',linestyle='--',label='Nc = 15',alpha=0.8)
ax[0].plot(dataSP_si[3][8:16],dataSP_si[3][0:8],marker='o',linestyle='--',label='Nc = 20',alpha=0.8)
ax[0].set_title('Tradeoff curves')
ax[0].set_xlabel('I(C;i) (bits)')
ax[0].set_ylabel('<S> (bits)')
ax[0].legend()

ax[1].plot(T_list,dataSP_si[0][16:24],marker='x',label='Nc = 5',alpha=0.8)
ax[1].plot(T_list,dataSP_si[1][16:24],marker='x',label='Nc = 10',alpha=0.8)
ax[1].plot(T_list,dataSP_si[2][16:24],marker='x',label='Nc = 15',alpha=0.8)
ax[1].plot(T_list,dataSP_si[3][16:24],marker='x',label='Nc = 20',alpha=0.8)
ax[1].set_title('deterministic curves')
ax[1].set_xlabel('1/T')
ax[1].set_ylabel('deterministic rate (%)')
ax[1].legend()
```
